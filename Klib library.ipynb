# Trying out the Data Cleaning Library Klib
This quick tutorial shows the capabilities of the Klib Library for Data Cleaning. I will be using the CCPP dataset previously used in predicting the output power of a combined cycle power plant 


## Missing Values Plot
Here we will be plot the missing values of the dataset 

```python
import klib
import pandas as pd

df = pd.read_csv(r'C:\Users\TIM\Desktop\AI Research Materials\Datasets\Regression\CCPP.csv')
```

Visualing the dataset for missing values

```python
klib.missingval_plot(df)
```

```python
df1 = pd.read_csv(r'C:\Users\TIM\Desktop\AI Research Materials\Datasets\Classification\credit_cards.csv')
klib.missingval_plot(df1)
```

```python
df2 = pd.read_csv(r'C:\Users\TIM\Desktop\AI Research Materials\Datasets\Regression\melb_data.csv')
klib.missingval_plot(df2)
```

## Dropping empty and duplicate rows and columns
The data_cleaning() functions essentially drops empty and single valued columns as well as empty and duplicate rows. We will implement that on df2 since it has missing values on attributes; Car, BuildingArea, YearBuilt, CouncilArea

```python
df2_cleaned = klib.data_cleaning(df2)
```

## Correlation Plot
This examines the intercorrelation amongst the features. 

```python
klib.corr_plot(df2_cleaned, annot=True)
```

```python
klib.corr_plot(df2, split= 'pos') #displays only positive correlation
klib.corr_plot(df2, split= 'neg') #display only negative correlation
```

```python
klib.dist_plot(df2)
```

```python
klib.cat_plot(df1)
```
